```python
batsman_runs = df_all.groupby('striker')['runs_of_bat'].sum().sort_values(ascending=False)
print(batsman_runs.head(10))
```

```python
wickets = df_all[df_all['wicket_type'].notna()] \
          .groupby('bowler')['wicket_type'].count() \
          .sort_values(ascending=False)

print(wickets.head(10))
```

```python
team_runs = df_all.groupby('batting_team')['runs_of_bat'].sum().sort_values(ascending=False)
print(team_runs)
```

```python
matches = df_all.groupby('batting_team')['match_id'].nunique()
print(matches)
```

```python
runs_over = df_all.groupby('over')['runs_of_bat'].sum()
print(runs_over)
```

```python
phase_runs = df_all.groupby('phase')['runs_of_bat'].sum()
print(phase_runs)
```
